## Start with the data

```python
# import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data
data_path = "./4-99-files/Lauren-store.csv"

store_data = pd.read_csv(data_path)

# examine the data
store_data.head(2)
```

### Plot the data using line chart

```python
# note the data is not clean, so don't take the results seriously
sns.lineplot(data=store_data)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(6, 5))

# Add title
plt.title("Lauren Store Data")

# Organize the data
store_data["revenue"] = store_data["revenue"].str.replace('$', '')
store_data["revenue"] = store_data["revenue"].str.replace(',', '')
store_data["revenue"] = store_data["revenue"].astype(float)

store_data["date"] = pd.to_datetime(store_data["date"])

# The line chart
sns.lineplot(data=store_data, x="date", y="revenue")
```

### Bar charts and heatmaps



```python
type(store_data["date"][1])
```

```python
# the same plot but with bar charts

plt.title("The revenu by color")

# data orgnization
store_data["date"] = pd.to_datetime(store_data["date"])

# the bar chart
sns.barplot(x=store_data["product_color"], y=store_data["revenue"])
```

```python
# let plot a heat map
plt.title("the heatmap")

data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
x_labels = ['A', 'B', 'C'] 
y_labels = ['D', 'E', 'F']

sns.heatmap(data, xticklabels=x_labels, yticklabels=y_labels)
```
