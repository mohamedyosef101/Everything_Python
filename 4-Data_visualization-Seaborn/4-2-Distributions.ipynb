```python
# load the libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data no.1

world_happiness = pd.read_csv("./4-99-files/world_happiness-2019.csv")
world_happiness.head(1)
```

```python
# load the data no.2
lauren_store = pd.read_csv("./4-99-files/Lauren-store.csv")
lauren_store.head(1)
```

```python
# Histogram
sns.histplot(data=world_happiness, x="Score")
```

### Density plots

```python
# KDE plot
sns.kdeplot(data=world_happiness, x='Score', fill=False)
```

```python
# 2D KDE plot
sns.jointplot(data=world_happiness, x="Score", y="Social support", kind='kde')
```

```python
# color-coded plots
plt.figure(figsize=(12, 10))

sns.histplot(data=lauren_store, x='product', hue='product_color')
```
