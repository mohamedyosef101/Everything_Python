<h2 id="range">Range</h2>


Before we discuss loops lets discuss the <span style="color: #00a1ff">range</span> object. It is helpful to think of the range object as an ordered list. For now, let's look at the simplest case. If we would like to generate an object that contains elements ordered from 0 to 2 we simply use the following command:


```python
# Use the range

range(3)
```

##### Some common ranges are:
- 65-90 (A-Z)
- 97-122 (a-z)
- 48-57 (0-9)
- 32-47 (symbols)

<h3>What is <span style="color: #00a1ff">for</span> loop?</h3>


```python
# For loop example

dates = [1982,1980,1973]
N = len(dates)

for i in range(N):
    print(dates[i])     
```

<h2 id="while">What is <span style="color: #00a1ff">while</span> loop?</h2>


**Here's an example of a while loop that prints numbers from 1 to 5:**


```python
count = 1
while count <= 5:
    print(count)
    count += 1
```

```python
# While Loop Example

dates = [1982, 1980, 1973, 2000]

i = 0
year = dates[0]

while(year != 1973):    
    print(year)
    i = i + 1
    year = dates[i]
    

print("It took ", i ,"repetitions to get out of loop.")
```
