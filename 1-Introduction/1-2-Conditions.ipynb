<h3 id="comp">Comparison Operators</h3>


Comparison operations compare some value or operand and based on a condition, produce a Boolean. When comparing two values you can use these operators:


| Operation | Symbol | Example |
| --- | --- | --- |
| Addition | + | 1 + 2 = 3 |
| Subtraction | - | 5 - 4 = 1 |
| Multiplication | * | 2 * 4 = 8 |
| Division | / | 6 / 3 = 2 |
| Exponent | ** | 3 ** 2 = 9 |


Let's assign <code>a</code> a value of 5. Use the equality operator denoted with two equal <b>==</b> signs to determine if two values are equal. The case below compares the variable <code>a</code> with 6.


```python
# Condition Equal

a = 5
a == 6
```

The result is <b>False</b>, as 5 does not equal to 6.


<h3 id="branch">Branching</h3>
The ability to execute different code blocks based on certain conditions in programming. 

<br>

<p><b>The main branching statements in Python are: </b></p>

- if
- if... else
- if... elif... else


```python
# If statement example

colors = ["red", "blue", "green"]
if "purple" in colors:
  print("purple is present")
```

```python
# Else statement example

age = 15

if age >= 18:
  print("You can vote")
else:
  print("You cannot vote")
```

```python
# Elif statment example

score = 70

if score >= 90:
  print("A grade")
elif score >= 80:
  print("B grade")  
elif score >= 70:
  print("C grade")
else:
  print("Fail")
```

<h3 id="logic">Logical operators</h3>


These allow building complex boolean logic in if statements and control flow.

<ul>
    <li><code>and</code> - True if both operands are true</li>
    <li><code>or</code> - True if either operand is true</li>
    <li><code>not</code> - True if operand is false (inverts boolean value)</li>
</ul>

```python
x = 5
y = 10

# and
if x > 0 and y > 5:
  print("Both conditions met")
  
# or  
if x > 0 or y < 0:
  print("One condition is true")
  
# not
if not x == y:
  print("x is not equal to y")
```
